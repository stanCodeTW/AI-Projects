```python
import os
from PIL import Image
import fitz
import pytesseract
import pdfplumber
from typing import List
```

```python
PDF_PATH = '/content/drive/MyDrive/Colab Notebooks/presentations/'
```

```python
def load(filepath) -> List[tuple]:
  files = []
  for root, dirs, files_in_dir in os.walk(filepath):
    for file_name in files_in_dir:
      if file_name.lower().endswith('.pdf'):
        file_path = os.path.join(root, file_name)
        files.append((file_path))
  return files
```

```python
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text("text")
    return text
```

```python
def extract_text_from_image_pdf(pdf_path):
  text = ""
  try:
    with pdfplumber.open(pdf_path) as pdf:
      for page in pdf.pages:
        pix = page.to_image()
        img = pix.original
        text += pytesseract.image_to_string(img)  # OCR extraction
  except Exception as e:
    print(f"Error extracting text from {pdf_path}: {e}")
  return text
```

```python
def load_and_extract_text_from_pdfs(pdf_path):
  pdf_files = load(pdf_path)
  all_texts = {}

  for pdf_file in pdf_files:
    print(f"Processing {pdf_file}...")

    # Try to extract text normally first (text-based PDFs)
    text = extract_text_from_pdf(pdf_file)

    # If no text was extracted, use OCR (image-based PDFs)
    if not text.strip():
      print(f"Using OCR for {pdf_file}...")
      text = extract_text_from_image_pdf(pdf_file)

    all_texts[os.path.basename(pdf_file)] = text

    return all_texts
```

```python
extracted_texts = load_and_extract_text_from_pdfs(PDF_PATH)
```
