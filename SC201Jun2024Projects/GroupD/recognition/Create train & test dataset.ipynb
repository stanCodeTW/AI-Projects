```python
# original folder path
source_folder = '../mirror_up'

# create train test folder
train_folder = './train'
val_folder = './val'
test_folder = './test'

# classes name
classes = ['Portrait', 'Stable Diffusion']

# 建立 train, val, test 資料夾
for folder in [train_folder, val_folder, test_folder]:
    for cls in classes:
        os.makedirs(os.path.join(folder, cls), exist_ok=True)
```

```python
# train/val/test ratio
train_split = 0.6
val_split = 0.2
test_split = 0.2

for cls in classes:
    cls_folder = os.path.join(source_folder, cls)
    images = os.listdir(cls_folder)
    random.shuffle(images)

    # 計算每個資料集的數量
    train_size = int(train_split * len(images))
    val_size = int(val_split * len(images))
    test_size = len(images) - train_size - val_size

    # 複製到對應的資料夾
    train_images = images[:train_size]
    val_images = images[train_size:train_size+val_size]
    test_images = images[train_size+val_size:]

    for image in train_images:
        source = os.path.join(cls_folder, image)
        destination = os.path.join(train_folder, cls, image)
        print(f"Copying {source} to {destination}")
        shutil.copy(source, destination)

    for image in val_images:
        source = os.path.join(cls_folder, image)
        destination = os.path.join(val_folder, cls, image)
        print(f"Copying {source} to {destination}")
        shutil.copy(source, destination)

    for image in test_images:
        source = os.path.join(cls_folder, image)
        destination = os.path.join(test_folder, cls, image)
        print(f"Copying {source} to {destination}")
        shutil.copy(source, destination)

print('Train data:' ,train_size)
print('Val data:' ,val_size)
print('Test data:' ,test_size)
```

```python
def count_images(folder_path):
    total_count = 0
    for root, dirs, files in os.walk(folder_path):
        image_count = len([f for f in files if f.endswith(('.png', '.jpg', '.jpeg'))])  # 根據你的圖片格式調整
        print(f"{root}: {image_count} images")
        total_count += image_count
    print(f"Total images in {folder_path}: {total_count}")
    return total_count

# calculate number of data in train, val, test
train_folder = 'train'
val_folder = 'val'
test_folder = 'test'

print("Train folder image counts:")
count_images(train_folder)

print("\nVal folder image counts:")
count_images(val_folder)

print("\nTest folder image counts:")
count_images(test_folder)
```

```python
# delete all images in these folder

def delete_images_in_folder(folder_path):
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # delete if it is image
            if file.endswith(('.png', '.jpg', '.jpeg')):
                file_path = os.path.join(root, file)
                # print(f"Deleting {file_path}")
                os.remove(file_path)

# define folders
train_folder = 'train'
val_folder = 'val'
test_folder = 'test'

delete_images_in_folder(train_folder)
delete_images_in_folder(val_folder)
delete_images_in_folder(test_folder)

print("All images deleted from train, val, and test folders.")
```
