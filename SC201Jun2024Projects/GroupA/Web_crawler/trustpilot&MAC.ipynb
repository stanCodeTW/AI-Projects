```python
import requests
from bs4 import BeautifulSoup
import csv
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
import re
comment_data = []
for page_num in range(1, 75):
    # Define the URL to scrape
    # print(page_num)
    url = f'https://www.trustpilot.com/review/www.apple.com?page={page_num}'
    # https://www.trustpilot.com/review/www.apple.com?page=2
    # ?page=2~5

    # Add a headers dictionary to simulate a browser visit
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all reviews using the appropriate class names from the file
        comments = soup.find_all('div', class_='styles_reviewCardInner__EwDq2')
        # print(comments)

        # with open('Trustpilot_commemt_data.csv', 'a', newline='', encoding='utf-8') as csvfile:
        with open('Trustpilot_comment_data.csv', 'a', newline='', encoding='utf-8') as csvfile:
            fieldnames = ['username', 'comment', 'date']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

            if page_num == 1:
                writer.writeheader()
            for comment in comments:

                comment_text = comment.find('div', class_='styles_reviewContent__0Q2Tg').text
                # print(comment_text)

                date = comment.find('time')
                if date:
                    date = date['datetime']
                else:
                    date = "Date not found"

                date_pattern = r'\b(January|February|March|April|May|June|July|August|September|October|November|December) \d{1,2}, \d{4}\b'
                date_of_experience = r'/b(Date_of_experience)'
                text_without_dates = []
                date_match = re.search(date_pattern, comment_text)

                if date_match:
                    date = date_match.group(0)
                    text_without_dates.append(re.sub(date_pattern,'', comment_text).strip())
                else:
                    text_without_dates.append(comment_text)

                comment_data.append({ "comment": text_without_dates, "time": date})

                print(f"Comment: {text_without_dates},{date}")

                writer.writerow({'comment': text_without_dates,'date':date})
```

```python
csv_file = "Trustpilot_comment_data.csv"
# /content/drive/My Drive/Colab\ Notebooks/Group_A_Project/
csv_file = "Trustpilot_comment_data.csv"
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["Reviews", "time"])
    writer.writeheader()
    for data in comment_data:
        writer.writerow(data)

print(f"Data has been saved to {csv_file}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
```

```python
chrome_options = Options()
chrome_options.binary_location = "/Applications/Google Chrome.app/Contents/MacOS/Google Chrome"
chrome_options.add_argument("--headless")  # 啟用無頭模式，不會打開瀏覽器視窗
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```python
# 初始化 WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# Apple Developer 論壇目標 URL
url = 'https://developer.apple.com/forums/thread/720003'

# 使用 Selenium 瀏覽頁面
driver.get(url)

# 等待頁面完全加載（可以根據網頁加載時間調整 sleep 時間）
time.sleep(3)  # 停留幾秒，確保 JavaScript 執行完成

# 獲取頁面 HTML 內容
page_source = driver.page_source

# 使用 BeautifulSoup 解析頁面
soup = BeautifulSoup(page_source, 'html.parser')

# 查找所有評論區塊
comments = soup.find_all('div', class_='chiclet-container')

# 保存評論
commemt_data_MAC = []

# 遍歷評論區塊
for comment in comments:
    # 提取標籤
    label = comment.find('a', class_='subtopic topic-kind')

    if label:
        label_text = label.text.strip()

        if label_text == "General":  # 僅抓取 General 分類
            print(22)

            # 提取評論內容
            comment_text = comment.find('div', class_='content')
            if comment_text:
                comment_text = comment_text.text.strip()
                commemt_data_MAC.append({"content": comment_text})
                print(comment_text)

# 關閉瀏覽器
driver.quit()

# 打印總共提取了多少評論
print(f"總共提取了 {len(commemt_data_MAC)} 條評論。")
```
