# 資料前處理

```python
# #label的資料為:  '對象', '國家', '地區', '景點類別'是label
trip_package = pd.read_excel('./tom_anchi_lion2.xlsx')
print(trip_package.columns)
trip_package_clean = trip_package[[
    '行程名稱', '行程總天數', '價格', '對象', '國家', '地區', '景點類別', '天', '景點數', '日程']]
trip_package_clean['天'] = trip_package_clean['天']+1
trip_package_clean['行程編號'] = trip_package_clean.groupby('行程名稱').ngroup() + 1
print(trip_package_clean.shape)
```

```python
# 把csv 行程一對多的doc變成一筆data--->這是for 有label的data
package_list=[]
# for name, group in trip_package_clean.groupby('行程名稱'):
for name, group in trip_package_clean.groupby('行程編號'):

    # print(name)
    # print(group['價格'].max())
    # 初始化行程内容字典
    itinerary_data = {
        # '行程名稱': name,
        '行程編號': name,
        '行程天數': int(group['行程總天數'].values[0]),
        '行程價格': group['價格'].values[0],
        '對象': group['對象'].values[0],
        '國家': group['國家'].values[0],
        '地區': group['地區'].values[0],
        '景點類別': group['景點類別'].values[0],
        '行程': {}
    }
    for day, data in group.iterrows():
        day_key = f"第{data['天']}天"
        itinerary_data['行程'][day_key] = []

        itinerary_data['行程'][day_key].append({
            "景點數": data['景點數'],
            "日程": data['日程']
        })
    # print(itinerary_data)
    # 将itinerary_data转换为JSON字符串
    itinerary_json = json.dumps(itinerary_data, ensure_ascii=False)
    tmp_df = pd.DataFrame(
        {'行程內容': [itinerary_json]})
    package_list.append(tmp_df)
package_df = pd.concat(package_list)
package_df.to_csv('package_df.csv', index=False,encoding='utf-8-sig')
```

# 用LLM（GPT-3.5 turbo）來生成回答

# 基本款：目的地行程推薦及注意事項

```python
# Build prompt
prompt_template = """
你是一名專業導遊，你的任務是組織各國的旅遊行程。\
此外，也為首次前往該國的旅客分享基本提示和注意事項。 \
如果你不知道答案，就說你不知道，不要試圖編造答案。 \
使用以下 上下文來回答最後的問題。 \
{context}
Question: {question}

# 請了解上下文後並依照以下格式回答:
關於...行程，我們推薦如下： :
推薦1:
第1天..
第2天..
第..天..
推薦2:
...
...

到這個國家需要注意的事情...例如從台灣到這個國家是否需要簽證, 這個國家的天氣, 這個國家的貨幣, 這個國家的風俗習慣等等，請列點說明:
1..
2..
...
"""

PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
)
# Run chain
question = f"推薦我情侶適合去的日本行程"
```

```python
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=label_vectordb.as_retriever(
    search_type="similarity", search_kwargs={'k': 5}), return_source_documents=True, verbose=True, chain_type_kwargs={"prompt": PROMPT})

result = qa_chain({"query": question})
print(result['result'])
```

```python
PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
)
# Run chain
question = input('請告訴我你想去的行程:')

qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=label_vectordb.as_retriever(
    search_type="similarity", search_kwargs={'k': 5}), return_source_documents=True, verbose=True, chain_type_kwargs={"prompt": PROMPT})

result = qa_chain({"query": question})
print(result['result'])
```
